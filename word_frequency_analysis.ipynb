## Positivity in Lyrics

##### Loading positive_words list and compare it to our lyrics

```python
positive_words = open("positive_words.txt","r")
positive_words_data = positive_words.read()
#replacing end of line with ' '
positive_words_list = positive_words_data.replace('\n', ' ').split(".")
# printing the data
#print(positive_words_list)
```

```python
# prueba loop
for line in positive_words_list:
    for word in line.split(" "):
        print(word)
```

```python
positive_words_per_song = []

for i in range (0, len(prepared_songs_dataset)):
    positive_words_found = []
    for line in positive_words_list:
        for positive_word in line.split(" "):
            if positive_word in prepared_songs_dataset.iloc[i]['Corpus']:
                print("Found positive word in record: {}".format(i), end="\r")
                positive_words_found.append(positive_word)
    positive_words_per_song.append(positive_words_found)
```

```python
prepared_songs_dataset['Positive Words'] = positive_words_per_song
prepared_songs_dataset.to_csv("final_prepped_dataset.csv")
```

```python
summary_positive_words = pd.DataFrame()
#years = prepared_songs_dataset['Year'].unique().tolist()

for i in range (0, len(years)):
    positive_words_per_year = sum(prepared_songs_dataset['Positive Words Counts'][prepared_songs_dataset['Year'] == years[i]])
    songs_with_positive_words_per_year = sum(prepared_songs_dataset['Positive Words Counts'][prepared_songs_dataset['Year'] == years[i]] > 3)

    row = {
        "Year": years[i],
        "Positive Words": positive_words_per_year,
        "Songs with Positive Words":songs_with_positive_words_per_year
    }
    summary_positive_words = summary_positive_words.append(row, ignore_index = True)
summary_positive_words['Year'] = summary_positive_words['Year'].astype(int)
summary_positive_words['Songs Per Year'] = characteristics['Ranking'].values.tolist()
summary_positive_words.set_index('Year', inplace=True)
```
